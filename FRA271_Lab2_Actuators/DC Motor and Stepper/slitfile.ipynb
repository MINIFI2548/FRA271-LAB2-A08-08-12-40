```python
import pandas as pd
import os
```

```python
# ระบุชื่อไฟล์ต้นฉบับ
file_path = 'test-3freqin-1file.xlsx'

# 2. อ่านไฟล์ Excel
# (จำเป็นต้องลง library openpyxl โดยพิมพ์ pip install openpyxl ใน terminal หากยังไม่มี)
try:
    df = pd.read_excel(file_path)
    print(f"อ่านไฟล์สำเร็จ: {file_path}")
except FileNotFoundError:
    print(f"ไม่พบไฟล์: {file_path} กรุณาตรวจสอบว่าไฟล์อยู่ในโฟลเดอร์เดียวกันหรือไม่")
    exit()
```

```python
# 2. ตรวจสอบชื่อคอลัมน์และดึงค่า Motor Frequency
target_col = 'Motor Frequency'
if target_col in df.columns:
    unique_freqs = df[target_col].unique()
    print(f"พบข้อมูลความถี่ (Motor Frequency) ทั้งหมด {len(unique_freqs)} ค่า: {unique_freqs}")
    
    # ดึงชื่อไฟล์และนามสกุลออกจากกัน (เพื่อเอาไปตั้งชื่อใหม่)
    file_name, file_extension = os.path.splitext(file_path)
    
    # 3. วนลูปสร้างไฟล์ใหม่ตามแต่ละความถี่
    for freq in unique_freqs:
        # เลือกข้อมูลเฉพาะความถี่นั้นๆ
        subset_df = df[df[target_col] == freq]
        
        # ตั้งชื่อไฟล์ใหม่: ชื่อเดิม_ความถี่.xlsx
        # เช่น test-3freqin-1file_1000.xlsx
        new_filename = f"{file_name}_{freq}{file_extension}"
        
        # บันทึกเป็นไฟล์ Excel
        subset_df.to_excel(new_filename, index=False)
        print(f" -> บันทึกไฟล์เรียบร้อย: {new_filename} (จำนวน {len(subset_df)} แถว)")
        
    print("\nเสร็จสิ้นกระบวนการทั้งหมด")

else:
    print(f"ไม่พบคอลัมน์ '{target_col}' ในไฟล์นี้ กรุณาตรวจสอบชื่อหัวตาราง")
```
